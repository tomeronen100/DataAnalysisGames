# **Q 4**

```python
import re

# Regex for url
r = re.compile("https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+")
# Find how many url has in every tweet
df["url-num"] = df["text"].apply(lambda txt: len(r.findall(txt)))
# Count the rows that have at least 1 url
num_of_tweet_with_url = len(df[df["url-num"] > 0])
rows_num = df.shape[0]

print(f"Percentage of tweets with url: {round(num_of_tweet_with_url/rows_num*100, 3)}%")
```

# **Q 5**

```python
# Regex for @ user mention
r = re.compile("\s*@(\w+)")
# Find user_mentions
df["user_mentions"] = df["text"].apply(lambda txt: r.findall(txt))

# Print all the unique user_mentions
user_mentions = []
for l in df["user_mentions"]:
  user_mentions += l

print(f"All the users mentions({len(user_mentions)}) from the tweets are: {user_mentions}"
      "\n\nAnd also created new column for user_mentions")
```

# **Q 6**

```python
# Regex for # user hashTags
r = re.compile("\s*#(\w+)")
# Find user_hashTags
df["user_hashTags"] = df["text"].apply(lambda txt: r.findall(txt))

# Print all the unique user_hashTags
user_hashTags = []
for l in df["user_hashTags"]:
  user_hashTags += l

print(f"All the users Hash Tags({len(user_hashTags)}) from the tweets are: {user_hashTags}"
      "\n\nAnd also created new column for user_hashTags")
```

```python
import nltk
nltk.download('stopwords')
```

```python
# Stop Words we will remove from entities
from nltk.corpus import stopwords
stop_Words = set(stopwords.words('english'))

# Regex for entities -
r = re.compile("[A-Z][A-Za-z]+")

# Find entities
df["entities"] = df["text"].apply(lambda txt: [word for word in r.findall(txt) if word.lower() not in stop_Words])

# Print all the unique entities
entities = []
for l in df["entities"]:
  entities += l

print(f"All the entities({len(entities)}) from the tweets are: {entities}"
      "\n\nAnd also created new column for entities")
```

# **Q 8**

```python
# Function that get a tweet and return list of numbers that has $ or % along them
#   Input: tweet: str
#   Output: list of numbers that has $ or % in the tweet
def getNumbersWithSign(tweet: str) -> list:
  numbers = []
  # Find numbers
  r = re.compile("[%$](\d*\.{0,1}\,{0,1}\d+)")
  numbers += r.findall(tweet)
  r = re.compile("(\d*\.{0,1}\,{0,1}\d+)[%$]")
  numbers += r.findall(tweet)

  return numbers

df["numbers_with_percents_or_dollars"] = df["text"].apply(getNumbersWithSign)
df.head()
```

# **Q 11**

```python
from wordcloud import WordCloud
# Function that get pandas series and visualize the words in it
#   Input: pd_series - Pandas Series that has lists of words
#   Output: Void - Create a Graph
def Visual_For_Words(pd_series):
  # Create a list of all the words
  words_txt = ""
  for words in pd_series:
    for word in words:
      words_txt = words_txt + word.lower() + " "
  # Visualize the words
  wordcloud = WordCloud(background_color='white', collocations=False).generate(words_txt)
  plt.figure(figsize = (15,15))
  plt.imshow(wordcloud)
  plt.axis('off')
```

```python
import pycountry
# Get countries list from pycountry Libary - Learn with Miki
countries_list_from_pycountry = [c.name for c in list(pycountry.countries)]
# Get the countries names from the df
countries_from_df = [country for country in df["tweet_location"]]

# Check that every country is real - if is: add +1 the count for that country
country_dict = {}
for country in countries_from_df:
  # Every country in the list we created
  for country2 in countries_list_from_pycountry:
    # Against every country in the pycountry list
    if(country.lower() in country2.lower() or country2.lower() in country.lower()):
      # Found match between two lists - Add the country
      name = country2
      if(name not in country_dict):
        country_dict[name] = 1
      else:
        country_dict[name] += 1

# Create DataFrame from the dictionary
countries_df = pd.DataFrame(list(country_dict.items()))
countries_df.columns = ["Name", "Number of Appearence"]

# Create Geo Graph from the DataFrame
fig = px.choropleth(countries_df, locations='Name', locationmode='country names',
                    color='Number of Appearence', color_continuous_scale='Viridis',
                    title='Number of Tweets on Earth')
fig.show()
```
