# **Marvel - חלק א**

## **Import The Marvel Universe Social Network: 'hero-network.csv' DB**

```python
df1 = pd.read_csv(MARVEL_PATH)
```

## **Q A.1**

```python
heros_friends = dict();
# Iter over all DB and create Dictionary of key:hero and values:list of friends
for index, row in df1.iterrows():
	if(row["hero1"] not in heros_friends):
		heros_friends[row["hero1"]] = set()
	if (row["hero2"] not in heros_friends):
		heros_friends[row["hero2"]] = set()
	heros_friends[row["hero1"]].add(row["hero2"])
	heros_friends[row["hero2"]].add(row["hero1"])

# Search for hero with one friend :(
heros_friends_with_one = [name for name in heros_friends if len(heros_friends[name]) == 1]
# Print Answer
print(f"Number of heroes with one friend: {len(heros_friends_with_one)}")
```

## **Q A.2**

```python
heros_friends = dict();
# Iter over all DB and create Dictionary of key:hero and values:list of friends
for index, row in df1.iterrows():
	if(row["hero1"] not in heros_friends):
		heros_friends[row["hero1"]] = set()
	if (row["hero2"] not in heros_friends):
		heros_friends[row["hero2"]] = set()
	heros_friends[row["hero1"]].add(row["hero2"])
	heros_friends[row["hero2"]].add(row["hero1"])

# Create 3 sets of every hero - will contain list of friends
friends_of_CaptainAmerica = set()
friends_of_SpiderMan = set()
friends_of_IronMan = set()

#  Iter over heroes friends dictionary and fill the sets
for name in heros_friends:
	if "captain america" in name.lower():
		friends_of_CaptainAmerica |= heros_friends[name]
	if "spider-man" in name.lower():
		friends_of_SpiderMan |= heros_friends[name]
	if "iron man" in name.lower():
		friends_of_IronMan |= heros_friends[name]

# Get the common friends of all 3
common_friends_Iron_Spider_Captain = friends_of_CaptainAmerica & friends_of_IronMan & friends_of_SpiderMan
# Print Answer
print(f"Number of common friends of Captain America, SpiderMan and IronMan: {len(common_friends_Iron_Spider_Captain)}")
```

## **Q A.3**

```python
# Create another column of len of name in df
df1["Len_Of_hero1"] = df1["hero1"].apply(len)
df1["Len_Of_hero2"] = df1["hero2"].apply(len)

# Get The max and min length of hero name
Small_Name_Size = min(df1[["Len_Of_hero1", "Len_Of_hero2"]].min())
Long_Name_Size = max(df1[["Len_Of_hero1", "Len_Of_hero2"]].max())

# Create set of all heros
all_heroes = set(df1["hero1"].values) | set(df1["hero2"].values)
# Sets will contain heroes with longest or shortest name
Small_Name_List = set()
Long_Name_List = set()

# Iter all_heroes Set and fill Small_Name_List and Long_Name_List Sets
for index, row in df1.iterrows():
  # SHort Name
  if(len(row["hero1"]) == Small_Name_Size):
    Small_Name_List.add(row["hero1"])
  if(len(row["hero2"]) == Small_Name_Size):
    Small_Name_List.add(row["hero2"])
    # Long Name
  if(len(row["hero1"]) == Long_Name_Size):
    Long_Name_List.add(row["hero1"])
  if(len(row["hero2"]) == Long_Name_Size):
    Long_Name_List.add(row["hero2"])

# Print Answer
print(f"Heroes with shortest Name: {Small_Name_List}")
print(f"Heroes with Longest Name: {Long_Name_List}")
```

## **Q A.4**

```python
# Function that check How many not common friends the given Hero Name has with Captain America
# Input:
#   str: hero name
# Return:
#   int: number of not common friends with Captain America
def Check_How_Many_Friends_Not_Common_With_CaptainAmerica(HeroName: str) -> int:
  # Set will contain the given HeroName friends and captain America
  HeroName_Friends = set()
  CaptainAmerica_Friends = set()
  for index, row in df1.iterrows():
    # Check for HeroName Friends
    if HeroName.lower() in row["hero1"].lower():
      HeroName_Friends.add(row["hero2"])
    if HeroName.lower() in row["hero2"].lower():
      HeroName_Friends.add(row["hero1"])
    # Check for Captain America Friends
    if "captain america" in row["hero1"].lower():
      CaptainAmerica_Friends.add(row["hero2"])
    if "captain america" in row["hero2"].lower():
      CaptainAmerica_Friends.add(row["hero1"])

  # Check how many friends the hero has
  num_of_friends_of_HeroName = len(HeroName_Friends)
  # Check how many common friends the hero and Captain America has
  num_of_common_friends_with_CaptainAmerica = len(set(HeroName_Friends & CaptainAmerica_Friends))
  # Return the diff - Not common friends value
  return num_of_friends_of_HeroName - num_of_common_friends_with_CaptainAmerica

# Call the function
num = Check_How_Many_Friends_Not_Common_With_CaptainAmerica("spider-man")
# Print Answer
print(f"Uncommon friends of given hero(spider-man) and Captain America is: {num}")
```

```python
df2 = pd.read_csv(EARTHQUAKES_PATH)
```

## **Q B.1**

```python
# Use Masking to get EarthQuakes with max Magnitude
strongest_earthquakes = df2[df2["Magnitude"] == df2["Magnitude"].max()]
# Print Answer
print(strongest_earthquakes)
```

## **Q B.2**

```python
# Use Masking to get the median of Nuclear Explosions
magnitude_of_nuclearExplosion = df2[df2["Type"] == "Nuclear Explosion"]["Magnitude"]
# Print Answer
print(f"The Median of the Magnitude of Earthquakes from Nuclear Explosion is: {magnitude_of_nuclearExplosion.median()}")
```

## **Q B.3**

```python
# Create new column of the Year
df2["Date"] = pd.to_datetime(df2["Date"], utc=True)
df2["Year"] = df2["Date"].dt.year

# Search for the max num of EarthQuakes in a year above Magnitude 5
magnitude_above_5 = df2[df2["Magnitude"] > 5]
magnitude_above_5_group_by_year = magnitude_above_5.groupby("Year")
max_earthquakes_inYear_above5 = int(magnitude_above_5_group_by_year.count()["Magnitude"].max())

# Set will contain the years that was the max num of EarthQuakes in a year above Magnitude 5
years_above_5 = set()

# Fill the Set
for year, num_of_earthquakes in magnitude_above_5_group_by_year.count()["Magnitude"].items():
  if(num_of_earthquakes == max_earthquakes_inYear_above5):
    years_above_5.add(year)

# Print Answer
print(f"The year that was maximum number earthqukes above 5 Magnitude is: {years_above_5}")

#############################################################################################

# Search for the max num of EarthQuakes in a year above Magnitude 8
magnitude_above_8 = df2[df2["Magnitude"] > 8]
magnitude_above_8_group_by_year = magnitude_above_8.groupby("Year")
max_earthquakes_inYear_above8 = int(magnitude_above_8_group_by_year.count()["Magnitude"].max())

# Set will contain the years that was the max num of EarthQuakes in a year above Magnitude 8
years_above_8 = set()

# Fill the Set
for year, num_of_earthquakes in magnitude_above_8_group_by_year.count()["Magnitude"].items():
  if(num_of_earthquakes == max_earthquakes_inYear_above8):
    years_above_8.add(year)

# Print Answer
print(f"The year that was maximum number earthqukes above 8 Magnitude is: {years_above_8}")
```

## **Q B.4**

```python
# Create new column of the Hour
df2["Hour"] = df2["Time"].apply((lambda time: time[:2]))
# Get the max number of EarthQuakes was at single Hour
group_by_hour = df2.groupby("Hour")
max_earthquakes_in_hour = group_by_hour.count().max()["Time"]

# Set will contain hours that was the maximum number of EarthQuakes at them
hours = set()

# Fill the Set
for hour, number in group_by_hour.count()["Time"].items():
  if(number == max_earthquakes_in_hour):
    hours.add(hour)

# Print Answer
print(f"The hour that has the most earthquakes is: {list(hours)}")
```

# **International football results - חלק ב**

## **Import International football results from 1872 to 2023: 'football_results.csv' DB**

```python
df3 = pd.read_csv(FOOTBALL_PATH)
```

## **Q C.1**

```python
# Function that receives as input the name of a team and returns the team that the given team
#   scores against her the most in away games in total
# Input:
#   str: team name
# Return:
#   set: teams that the given team score against her the most in away games in total
def Most_Conqueror_Against(team: str) -> set:
  # Get only the games of team
  games_of_team = df3[df3["away_team"] == team]
  # Group by and get the max of sum score at away games
  max_score_of_team_at_AwayGames = games_of_team.groupby("home_team")["away_score"].sum().max()
  # Set will contain the Home Teams
  home_teams = set()
  for team, score in games_of_team.groupby("home_team")["away_score"].sum().items():
    if(score == max_score_of_team_at_AwayGames):
      home_teams.add(team)
  # Return the teams
  return home_teams

# Print Answer
# Italy:
print(f"Italy most conqueror against: {Most_Conqueror_Against('Italy')}")
# Argentina
print(f"Argentina most conqueror against: {Most_Conqueror_Against('Argentina')}")
# Israel
print(f"Israel most conqueror against: {Most_Conqueror_Against('Israel')}")
```

## **Q C.2**

```python
# A function that get 2 teams names and returns the date of the game and the result of
# the game with the most goals, and the game with the fewest goals, splited by home game and away game
# Input:
#   str: team1 name
#   str: team2 name
# Return:
#   List: List of Lists that:
#                 List[0] - team1 Home & team2 Away: most goals
#                 List[1] - team1 Home & team2 Away: fewest goals
#                 List[2] - team2 Home & team1 Away: most goals
#                 List[3] - team2 Home & team1 Away: fewest goals
def most_goals_least_score_HomeOrAway(team1: str, team2: str) -> list:
  games = []

  # Filter team1 Home & team2 Away and add column of Total Score
  home_1_away_2 = df3[(df3["home_team"] == team1) & (df3["away_team"] == team2)].copy()
  home_1_away_2["TotalScore"] = home_1_away_2["home_score"] + home_1_away_2["away_score"]
  # Get max Score and min score
  home_1_away_2_max = home_1_away_2["TotalScore"].max()
  home_1_away_2_min = home_1_away_2["TotalScore"].min()
  # Get games with that score
  home_1_away_2_max_games = home_1_away_2[home_1_away_2["TotalScore"] == home_1_away_2_max]
  home_1_away_2_min_games = home_1_away_2[home_1_away_2["TotalScore"] == home_1_away_2_min]
  # Get the max latest match
  home_1_away_2_max_games = home_1_away_2_max_games[home_1_away_2_max_games["date"] == max(home_1_away_2_max_games["date"])]
  home_1_away_2_min_games = home_1_away_2_min_games[home_1_away_2_min_games["date"] == max(home_1_away_2_min_games["date"])]
  # Add them to the return list
  games.append(home_1_away_2_max_games[["date", "home_score", "away_score"]])
  games.append(home_1_away_2_min_games[["date", "home_score", "away_score"]])

  # Filter team1 Away & team2 Home and add column of Total Score
  away_1_home_2 = df3[(df3["away_team"] == team1) & (df3["home_team"] == team2)].copy()
  away_1_home_2["TotalScore"] = away_1_home_2["home_score"] + away_1_home_2["away_score"]
  # Get max Score and min score
  away_1_home_2_max = away_1_home_2["TotalScore"].max()
  away_1_home_2_min = away_1_home_2["TotalScore"].min()
  # Get games with that score
  away_1_home_2_max_games = away_1_home_2[away_1_home_2["TotalScore"] == away_1_home_2_max]
  away_1_home_2_min_games = away_1_home_2[away_1_home_2["TotalScore"] == away_1_home_2_min]
  # Get the max latest match
  away_1_home_2_max_games = away_1_home_2_max_games[away_1_home_2_max_games["date"] == max(away_1_home_2_max_games["date"])]
  away_1_home_2_min_games = away_1_home_2_min_games[away_1_home_2_min_games["date"] == max(away_1_home_2_min_games["date"])]
  # Add them to the return list
  games.append(away_1_home_2_max_games[["date", "home_score", "away_score"]])
  games.append(away_1_home_2_min_games[["date", "home_score", "away_score"]])

  return games

# Example run Spaing and Germany
team1 = "Spain"
team2 = "Germany"
games = most_goals_least_score_HomeOrAway(team1, team2)

print(f"####### {team1} is the Home Team, {team2} is the Away Team #######")
print(f"## The game with the greatest goals is:\n{games[0]}\n")
print(f"## The game with the fewest goals is:\n{games[1]}\n\n")

print(f"####### {team2} is the Home Team, {team1} is the Away Team #######")
print(f"## The game with the greatest goals is:\n{games[2]}\n")
print(f"## The game with the fewest goals is:\n{games[3]}")
```
